Matplotlib

```python
partner_counts = df["delivery_partner"].value_counts()

plt.figure()
partner_counts.plot(kind="bar")

plt.title("Count of Deliveries by Delivery Partner")
plt.xlabel("Delivery Partner")
plt.ylabel("Number of Deliveries")

plt.show()
```

```python
plt.figure()

df["delivery_id"].plot(kind="hist", bins=30)

plt.title("Distribution of Delivery ID")
plt.xlabel("Delivery ID")
plt.ylabel("Frequency")

plt.show()
```

```python
missing = df.isnull().sum()

plt.figure()

missing.plot(kind="bar")

plt.title("Missing Values per Column")
plt.xlabel("Columns")
plt.ylabel("Number of Missing Values")

plt.show()
```

```python
df = df.drop(columns=["Unnamed: 15","Unnamed: 16","Unnamed: 17"])
```

```python
plt.figure()

plt.scatter(df["distance_km"], df["delivery_time_hours"])

plt.xlabel("Distance (KM)")
plt.ylabel("Delivery Time (Hours)")
plt.title("Distance vs Delivery Time")

plt.show()
```

```python
plt.figure()

df["delivery_cost"].plot(kind="hist", bins=25)

plt.title("Delivery Cost Distribution")
plt.xlabel("Delivery Cost")
plt.ylabel("Frequency")

plt.show()
```

```python
vehicle_counts = df["vehicle_type"].value_counts()

plt.figure()

vehicle_counts.plot(kind="bar")

plt.title("Vehicle Type Used for Delivery")
plt.xlabel("Vehicle Type")
plt.ylabel("Count")

plt.show()
```

```python
region_counts = df["region"].value_counts()

plt.figure()

region_counts.plot(kind="bar")

plt.title("Deliveries by Region")
plt.xlabel("Region")
plt.ylabel("Number of Deliveries")

plt.show()
```
